```python
import pandas as pd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv("./stsim_m_top10_maha.txt", header=None)
df = pd.read_csv("./stsim_2_global_result.txt", header=None)
df = df.iloc[:,:11]
df.columns=["query_img","top1","top2","top3","top4","top5","top6","top7","top8","top9","top10"]
df.head()
```

```python
data_dir = "/Users/yuxiao/Desktop/data/Corbis128BigExperiment_gray/"
```

```python
def show_img(df, n):
    columns = 4
    rows = int(np.ceil(10+1/columns))
    fig = plt.figure(figsize=(20,50))
    col_len = len(df.iloc[0])
    for i in range(col_len):
        image = Image.open(data_dir + df.iloc[n,i])
        fig.add_subplot(rows, columns, i+1)
        plt.gca().set_title("%s"%df.iloc[n,i],size=16)
        plt.imshow(image,cmap='gray')
    return plt.show()
```

```python
show_img(df,12)
```
